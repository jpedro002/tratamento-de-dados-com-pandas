```python
import pandas as pd

df = pd.read_excel('Respostas.xlsx')

display(df)
```

```python
mapeamento = {
    'Não costumo jogar' : 0,
    'Menos de 30 Minutos': 0,
    'Menos de 1 Hora': 0,
    '1 Hora': 1,
    '2 Horas': 2,
    '3 Horas': 3,
    '4 Horas': 4,
    '5 Horas': 5,
    '5+ Horas': 5,
}

df['Aproximadamente quanto tempo você dedica aos estudos da instituição por dia?'] = df['Aproximadamente quanto tempo você dedica aos estudos da instituição por dia?'].replace(mapeamento)
df['Aproximadamente quanto tempo você dedica aos jogos por dia?'] = df['Aproximadamente quanto tempo você dedica aos jogos por dia?'].replace(mapeamento)
```

```python
display(df)
#[['Aproximadamente quanto tempo você dedica aos estudos por dia?','Aproximadamente quanto tempo você dedica aos jogos por dia?']]
```

```python

ocorrencia_de_dispositivos = df[['Quais dessas plataformas você mais utiliza?','Aproximadamente quanto tempo você dedica aos estudos da instituição por dia?','Aproximadamente quanto tempo você dedica aos jogos por dia?']].groupby('Quais dessas plataformas você mais utiliza?').value_counts()

display(ocorrencia_de_dispositivos)

```

```python
import matplotlib.pyplot as plt

# Contagem de Observações por Categoria
contagem_plataformas = df['Quais dessas plataformas você mais utiliza?'].value_counts()

# Estatísticas Descritivas
estatisticas_estudos = df['Aproximadamente quanto tempo você dedica aos estudos da instituição por dia?'].describe()
estatisticas_jogos = df['Aproximadamente quanto tempo você dedica aos jogos por dia?'].describe()

# Análise de Correlação
correlacao = df['Aproximadamente quanto tempo você dedica aos estudos da instituição por dia?'].corr(df['Aproximadamente quanto tempo você dedica aos jogos por dia?'])

print(estatisticas_estudos)

# Visualizações Gráficas
df['Quais dessas plataformas você mais utiliza?'].value_counts().plot(kind='bar')
plt.title('Contagem de Observações por Plataforma')
plt.xlabel('Quais dessas plataformas você mais utiliza?')
plt.ylabel('Contagem')
plt.show()
```

```python
df.groupby('Quais dessas plataformas você mais utiliza?')[['Aproximadamente quanto tempo você dedica aos estudos da instituição por dia?','Aproximadamente quanto tempo você dedica aos jogos por dia?']].mean()
```

```python
import numpy as np

media = df.groupby('Quais dessas plataformas você mais utiliza?')[['Aproximadamente quanto tempo você dedica aos estudos da instituição por dia?', 'Aproximadamente quanto tempo você dedica aos jogos por dia?']].mean()

# Criar o gráfico
fig, ax = plt.subplots(figsize=(30, 7))
x = np.arange(len(media.index))
bar_width = 0.20

b1 = ax.bar(x - bar_width/2, media['Aproximadamente quanto tempo você dedica aos estudos da instituição por dia?'], width=bar_width, label='Estudo')
b2 = ax.bar(x + bar_width/2, media['Aproximadamente quanto tempo você dedica aos jogos por dia?'], width=bar_width, label='Jogo')

# Adicionar rótulos aos grupos de barras
ax.set_xticks(x)
ax.set_xticklabels(media.index)
ax.legend()

plt.ylabel('Média de horas')
plt.title('Média de horas de estudo e jogo por plataforma')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

mapeamento_faixa_etaria = {
    '18 a 20': '18 a 30',
    '21 a 24' : '18 a 30',
    '25 a 29': '18 a 30',
    '30 a 40': '30 a 50',
    '40 a 50':'30 a 50',
    '50': '50+'
}

df['Sua faixa etária:'] = df['Sua faixa etária:'].replace(mapeamento_faixa_etaria)


faixas_etarias = ['18 a 30', '30 a 50', '50+']

for idade in faixas_etarias:
    # Filtrar os dados para a faixa etária específica
    data_por_idade = df[df['Sua faixa etária:'] == idade]

    # Calcular a média para estudo e jogo
    media = data_por_idade.groupby('Quais dessas plataformas você mais utiliza?')[['Aproximadamente quanto tempo você dedica aos estudos da instituição por dia?', 'Aproximadamente quanto tempo você dedica aos jogos por dia?']].mean()

    # Criar o gráfico
    fig, ax = plt.subplots(figsize=(20, 7))
    x = np.arange(len(media.index))
    bar_width = 0.35

    b1 = ax.bar(x - bar_width/2, media['Aproximadamente quanto tempo você dedica aos estudos da instituição por dia?'], width=bar_width, label='Estudo')
    b2 = ax.bar(x + bar_width/2, media['Aproximadamente quanto tempo você dedica aos jogos por dia?'], width=bar_width, label='Jogo')

    
    # Adicionar rótulos aos grupos de barras
    ax.set_xticks(x)
    ax.set_xticklabels(media.index)
    ax.legend()

    plt.ylabel('Média de horas')
    plt.title(f'Média de horas de estudo e jogo por plataforma - Faixa Etária: {idade}')
    plt.show()
```
